# Question 1

## What is the cardinality in “basket data.csv”?

```python
def cardinality_items(filename):
  try:
    with open(filename, "r") as file:
      content = file.read()

    unique_items = set()

    lines = content.split("\n")

    for line in lines:
      if line:
        items = line.split(",")
        for item in items:
          unique_items.add(item.strip())
    return len(unique_items)

  except FileNotFoundError:
    print(f"File '{filename}' not found.")
    return 0

cardinality = cardinality_items("basket_data.csv")
print(f"Cardinality in 'basket_data.csv' is {cardinality}")
```

# Question 3

## (a) Total records of movie ratings in the entire dataset

```python
def total_ratings():
  total = 0
  for i in range(1, 5):
    filename = f'combined_data_{i}.txt'
    with open(filename, "r") as file:
      ratings = file.read()

    for line in ratings.split("\n"):
      if line:
        if line.endswith(":"):
          continue
        total += 1
  return total
```

```python
print(total_ratings())
```

## (b) Total Unique Users

```python
def total_unique_users():
  unique_users = set()
  for i in range(1, 5):
    filename = f'combined_data_{i}.txt'
    with open(filename, "r") as file:
      combined_data = file.readlines()

    for data in combined_data:
      data = data.rstrip()
      if data and not data.endswith(":"):
        user_id = data.split(",")[0]
        unique_users.add(user_id)
  return len(unique_users)
```

```python
print(total_unique_users())
```

## (c) What range of year that this data is valid over?

```python
def date():

  for i in range(1, 5):
      filename = f'combined_data_{i}.txt'
      with open(filename, "r") as file:
        date = file.read()

      all_year = set()

      for data in date.split("\n"):
        if data.endswith(":"):
          continue
        if len(data) >= 3:
          year = data.split(",")[2].split("-")[0]
          all_year.add(int(year))
  return [min(all_year), max(all_year)]
```

```python
print("Range of years: ", date()[0], "to", date()[1])
```

## (a) How many movies with unique names are there?

```python
import csv
def unique_movies(filename):
  unique = set()
  duplicate = set()

  with open(filename, "r", encoding="cp1252") as file:
    lines = csv.reader(file)

    for line in lines:
      if len(line) >= 3:
        movie = " ".join(line[2:]).strip()
        if movie in unique:
          duplicate.add(movie)
        else:
          unique.add(movie)
  print(unique)
  return len(unique)

unique_movies("movie_titles.csv")
```

# Question 5

## (a) How many users rated exactly 200 movies?

```python
def count_exact():
  count = {}

  for file_number in range(1, 5):
    filename = f'combined_data_{file_number}.txt'
    with open(filename, "r") as file:
      line = file.read()
    for data in line.split("\n"):
      if data and not data.endswith(":"):
        customer = int(data.split(",")[0])
        if customer not in count:
          count[customer] = 1
        else:
          count[customer] += 1
  return count
```

```python
def exact_movies(count):
  res = 0
  for c in count.values():
    if c == 200:
      res += 1
  return res
```

```python
total = exact_movies(count_exact())
print(total)
```

## (b) Of these users, take the lowest user ID and print out the names of the movies that they liked the most (all 5 star ratings).

```python
def lowest_user_id_in_200(count):
  users = set()
  for customer_id, count in count.items():
    if count == 200:
      users.add(customer_id)
  return min(users)
```

```python
from collections import defaultdict

def movie_id_list(user):
  movie_id_list = defaultdict(set)

  for file_number in range(1, 5):
    filename = f'combined_data_{file_number}.txt'
    with open(filename, "r") as file:
      lines = file.readlines()
    for line in lines:
      line = line.rstrip()
      if line and line.endswith(":"):
          movie_id = int(line.split(":")[0])
      if line and len(line.split(",")) >= 3:
          customer = int(line.split(",")[0])
          rating = int(line.split(",")[1])
          if rating == 5:
            movie_id_list[customer].add(movie_id)
  return movie_id_list[user]
```

```python
from collections import defaultdict
def most_liked(movie_id_list, movie_file):
  most_liked = set()
  with open(movie_file, "r", encoding="cp1252") as file:
    lines = file.readlines()
  for line in lines:
    line = line.rstrip()
    if line and len(line.split(",")) >= 3:
      movie_id = int(line.split(",")[0])
      if movie_id in movie_id_list:
        title = "".join(line.split(",")[2:])
        most_liked.add(title)
  return most_liked
```

```python
lowest_user_id = lowest_user_id_in_200(count_exact())
lowest_user_id
```

```python
movie_list = movie_id_list(lowest_user_id)
```
